```python
# 1. prompt the user to enter their personal information such as name, age, address, salary, 
#and expenses

name= input("Enter your name:")
age= int(input("Enter your age in numbers:"))
address=input("Enter your address:")
salary= round(float(input("Enter your salary in numbers:")),1)
expenses= round(float(input("Enter the total amount of your expenses in numbers:")),1)
remaining_salary=salary-expenses


```

```python
#Use a boolean variable to indicate whether the remaining salary is greater than or equal to 500.
check= remaining_salary>=500
print(f"Hey {name}, who is {age} years old, and lives in {address} has {remaining_salary} dollars left from her salary after expenses. It is {check} that she has more than $500 left.It is {check} that the money you have left is >=500")

```
